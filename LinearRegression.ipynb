```python
areas = [1000, 5000, 10000, 15000]
```

```python
prices = [800, 5200, 10500, 14000]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(areas, prices)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
import numpy as np
areas_2d = np.array(areas).reshape(-1, 1)
model.fit(areas_2d, prices)
```

```python
prediction = model.predict([[5000]])
prediction
```

```python
model.coef_
```
